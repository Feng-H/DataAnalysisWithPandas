```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['font.sans-serif'] = 'SimHei'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
io_exchange_rates = './exchange_rate_data/forex.csv'
df_exchange_rate = pd.read_csv(io_exchange_rates)
```

```python
df_exchange_rate.head()
```

```python
df_CNY = df_exchange_rate[df_exchange_rate['slug']=='USD/CNY']
df_CNY.head()
```

```python
df_CNY.info()
```

```python
df_CNY.describe()
```

```python
df_CNY.plot(x='date',title="美元对人民币汇率",figsize=(16,16),subplots=True)
```

```python
df_CNY['open_diff']=df_CNY['open'].diff()
df_CNY['high_diff']=df_CNY['high'].diff()
df_CNY['low_diff']=df_CNY['low'].diff()
df_CNY['close_diff']=df_CNY['close'].diff()
```

```python
df_CNY
```

```python
df_CNY.plot(x='date',title='美元对人民币汇率',figsize=(16,16),subplots=True) 
```

```python
df_CNY[(df_CNY['open_diff']<0.5) * (df_CNY['open_diff']> -0.5) * 
       (df_CNY['high_diff']<0.5) * (df_CNY['high_diff']> -0.5) * 
       (df_CNY['low_diff']<0.5) * (df_CNY['low_diff']> -0.5) *
       (df_CNY['close_diff']<0.5) * (df_CNY['close_diff']> -0.5)].plot(x='date',title='USD to RMB exchange rate',figsize=(16,16),subplots=True)
```

```python
a = 0.57
```

```python
df_CNY[(df_CNY['open_diff']< a) * (df_CNY['open_diff']> -a) * 
       (df_CNY['high_diff']< a) * (df_CNY['high_diff']> -a) * 
       (df_CNY['low_diff']< a) * (df_CNY['low_diff']> -a) *
       (df_CNY['close_diff']< a) * (df_CNY['close_diff']> -a)].plot(x='date',title='USD to RMB exchange rate',figsize=(16,16),subplots=True)
```
